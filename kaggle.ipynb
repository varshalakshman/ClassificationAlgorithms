```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing dataset
col_names = [i for i in range(101)]
dataset = pd.read_csv("train_features.csv",names = col_names)

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
# Removing first column
X_train = dataset.iloc[:, 1: ].values
```

```python
# Importing Training Labels
col_names1 = [i for i in range(2)]
labels = pd.read_csv("train_label.csv",names = col_names1)
y_train = labels.iloc[1:, -1].values
print(X_train.shape)
print(y_train.shape)
```

```python
# Importing test features
temp1 = [i for i in range(101)]
test_data = pd.read_csv("test_features.csv",names = temp1)
X_test = test_data.iloc[:, 1: ].values
print(X_test.shape)
```

```python
# Doing scalarization of training and test data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Implementing SVM
from sklearn.svm import SVC
svcl = SVC(kernel='poly', gamma=9, degree=2)
svcl.fit(X_train, y_train)
y_pred_svm = svcl.predict(X_test)
y_pred_svm = y_pred_svm.astype(int)
print(y_pred_svm)
print(type(y_pred_svm))
y_pred_svm = y_pred_svm.tolist()
print(y_pred_svm)
print(type(y_pred_svm))
```

```python
#Implementing KNN
from sklearn.neighbors import KNeighborsClassifier
kcl = KNeighborsClassifier(n_neighbors=6, weights='distance')
# print(kcl)
kcl.fit(X_train, y_train)
y_pred_knn = kcl.predict(X_test)
# print(y_pred_knn)
y_pred_knn = y_pred_knn.astype(int)
print(y_pred_knn)
y_pred_knn = y_pred_knn.tolist()
print(y_pred_knn)
print(type(y_pred_knn))
```

```python
#Implementing Logistic Regression
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=10, multi_class='ovr', C=0.1, penalty='l1')
# print(LR)
lrc.fit(X_train,y_train)
y_pred_lr = lrc.predict(X_test)
y_pred_lr = y_pred_lr.astype(int)
print(y_pred_lr)
y_pred_lr = y_pred_lr.tolist()
print(y_pred_lr)
print(type(y_pred_lr))
print(len(y_pred_lr))
```

```python
#Implementing Random Forest
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=20, random_state=8, max_features=100)
regr.fit(X_train, y_train)
y_pred1 = regr.predict(X_test)
print(y_pred1)
print(type(y_pred1))

y_pred_rf = []
for item in y_pred1:
    temp = round(item)
    y_pred_rf.append(temp)

y_pred_rf = np.asarray(y_pred_rf)
y_pred_rf = y_pred_rf.astype(int)
print(y_pred_rf)

y_pred_rf = y_pred_rf.tolist()
print(y_pred_rf)
print(type(y_pred_rf))
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Create adaboost classifer object
abc =AdaBoostClassifier(n_estimators=9, learning_rate=0.5)

print(abc)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred_ada = model.predict(X_test)
y_pred_ada = y_pred_ada.astype(int)
print(y_pred_ada)

y_pred_ada = y_pred_ada.tolist()
print(y_pred_ada)
print(type(y_pred_ada))
```

```python
df = pd.DataFrame(list(zip(y_pred_svm, y_pred_lr, y_pred_knn, y_pred_rf, y_pred_ada)), index =[i for i in range(418,796)])
```
